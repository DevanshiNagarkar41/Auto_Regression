```python
import random
from datetime import datetime, timedelta

# Function to generate a random date within a range
def generate_random_date(start, end):
    delta = end - start
    return start + timedelta(days=random.randint(0, delta.days))

# Define the date range
start = datetime(1960, 1, 1)
end = datetime(2000, 12, 31)

# Load the CSV file
file_location = 'C:/Users/admin/Downloads/data (1).csv'
data = pd.read_csv(file_location)

# Fill missing 'DOB' values with random dates
data['DOB'] = data['DOB'].apply(lambda date: generate_random_date(start, end) if pd.isnull(date) else date)

data
```

```python
import pandas as pd
import json
import os

# Existing symptoms dictionary
symptoms_dict = {
    'S1': {'Fever': ['mild', 'low', 'high']},
    'S2': {'Cough': ['mild', 'low', 'high']},
    'S3': {'Cold': ['mild', 'low', 'high']},
    'Body Ache': {'Severity': ['mild', 'low', 'high']},
    'Shivering': {'Severity': ['mild', 'low', 'high']},
    'Head Ache': {'Severity': ['mild', 'low', 'high']},
    'Nausea': {'Severity': ['mild', 'low', 'high']},
    'Sickness': {'Severity': ['mild', 'low', 'high']},
    'Vertigo': {'Severity': ['mild', 'low', 'high']}
}

# Function to read data from various file formats
def load_data(file_path):
    _, file_extension = os.path.splitext(file_path)
    
    if file_extension == '.csv':
        return pd.read_csv(file_path)
    elif file_extension == '.tsv':
        return pd.read_csv(file_path, sep='\t')
    elif file_extension == '.json':
        return pd.read_json(file_path)
    elif file_extension == '.xml':
        return pd.read_xml(file_path)
    else:
        raise ValueError("Unsupported file format")

# Function to update the dictionary with new symptoms
def update_dictionary(df):
    added_symptoms = []

    # Extract symptoms from the 'Other_Symptoms' column
    for index, row in df.iterrows():
        other_symptoms = row['Other_Symptoms'].split(',')
        
        for symptom in other_symptoms:
            symptom_key = symptom.strip()  # Clean leading/trailing spaces
            if symptom_key not in symptoms_dict:
                symptoms_dict[symptom_key] = {"Severity": ["mild", "low", "high"]}
                added_symptoms.append(symptom_key)

    return added_symptoms

# Function to save the dictionary to a JSON file
def save_dictionary(file_name):
    with open(file_name, 'w') as file:
        json.dump(symptoms_dict, file, indent=4)

# Function to load the dictionary from a JSON file
def load_symptoms_from_file(file_name):
    global symptoms_dict
    with open(file_name, 'r') as file:
        symptoms_dict = json.load(file)

# Function to print the symptoms dictionary
def print_dictionary():
    print("=== Current Symptoms Dictionary ===")
    print(json.dumps(symptoms_dict, indent=4))

# Function to manually add new symptoms
def manual_symptom_entry():
    while True:
        symptom = input("Enter a symptom to add (or type 'exit' to stop): ").strip()
        if symptom.lower() == 'exit':
            break
        
        if symptom not in symptoms_dict:
            severity = input("Enter severity levels (comma-separated, e.g., mild, low, high): ").strip().split(',')
            severity = [s.strip() for s in severity]  # Clean spaces
            symptoms_dict[symptom] = {"Severity": severity}
            print(f"Symptom '{symptom}' added with severity levels: {severity}.")
        else:
            print(f"Symptom '{symptom}' already exists in the dictionary.")

# Main function to execute the program
def main():
    # Example file paths (update these paths as needed)
    file_path = 'data.csv'  # Example CSV file path

    # Step 1: Load data
    try:
        df = load_data(file_path)
    except Exception as e:
        print(f"Error loading file: {e}")
        return

    # Step 2: Update the dictionary with new symptoms
    added_symptoms = update_dictionary(df)

    # Step 3: Print the updated dictionary
    print_dictionary()

    # Print newly added symptoms
    if added_symptoms:
        print("\nNewly added symptoms:")
        for symptom in added_symptoms:
            print(symptom)
    else:
        print("\nNo new symptoms were added.")

    # Step 4: Allow manual addition of symptoms
    manual_symptom_entry()

    # Step 5: Save the enhanced dictionary to a JSON file
    save_dictionary('enhanced_symptoms_dict.json')

# Execute the main function
if __name__ == "__main__":
    main()
```
